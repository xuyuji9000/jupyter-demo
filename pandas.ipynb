```python
import pandas as pd
```

```python
# dataframe 
raw_data = {'first_name': ['Jason', 'Molly', 'Tina', 'Jake', 'Amy'], 
        'last_name': ['Miller', 'Jacobson', ".", 'Milner', 'Cooze'], 
        'age': [42, 52, 36, 24, 73], 
        'preTestScore': [4, 24, 31, ".", "."],
        'postTestScore': ["25,000", "94,000", 57, 62, 70]}
df = pd.DataFrame(raw_data, columns = ['first_name', 'last_name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
df.to_csv('example.csv')
```

```python
# load from csv
df = pd.read_csv('example.csv', header=None)
df
```

```python
# load with colome name
df = pd.read_csv('example.csv', names = ['first_name', 'last_name', 'age', 'preTestScore', 'postTestScore'])
df
```

```python
# load with index column
df = pd.read_csv('example.csv', index_col=['UID'], names=['UID', 'First Name', 'Last Name', 'Age', 'Pre-Test Score', 'Post-Test Score'])
df
```

```python
# load . as missing values
df = pd.read_csv('example.csv', na_values=['.', 'NA'])
# pd.isnull(df)
df
```

```python
# make NaN
sentinels = {'last_name': ['.', 'NA', 'Miller'], 'preTestScore': ['.']}
df = pd.read_csv('example.csv', na_values=sentinels)
df
```

```python
# skip top 3 rows
df = pd.read_csv('example.csv', skiprows=3)
df
```

```python
# sort by age 
df = pd.read_csv('example.csv')
df.sort_values(by=['age'])
```
