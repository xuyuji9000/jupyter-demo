```python
import pandas as pd 
import matplotlib.pyplot as plt
import descartes
import geopandas as gpd

%matplotlib inline
```

```python


street_map = gpd.read_file("./shapfiles/nyct2010_15b")
```

```python
# street_map = street_map.to_crs(crs="+proj=longlat +datum=WGS84")
street_map = street_map.to_crs(crs={'proj': 'latlong', 'datum': 'WGS84'})
map_ex_staten_island = street_map[street_map['BoroName'] != 'Staten Island']
```

```python
df = pd.read_csv('./uber-raw-data-apr14.csv')
df.columns
gdf = gpd.GeoDataFrame(df, geometry=gpd.points_from_xy(df.Lon, df.Lat))
```

```python
fig,ax = plt.subplots(figsize = (50, 50))
ax.set_facecolor("#080808")
base = map_ex_staten_island.plot(ax=ax, alpha=0.14)

gdf[1:1000].plot(ax=base, marker='o', color='red', markersize=1);

plt.savefig('test.png')
# gdf.plot(ax=base, marker='o', color='red', markersize=1);
# type(map_ex_staten_island)
# map_ex_staten_island.head()
```
