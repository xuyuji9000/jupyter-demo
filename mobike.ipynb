```python
# import data
import pandas as pd
df = pd.read_csv('./data/UTSEUS-MOBIKE-shanghai_full.csv')
```

```python
# get one user data
first_user = df.query("userid == 7541")
first_user.sort_values(by=['start_time'])
```

```python
# get a track
track = first_user['track'][27274].split('#')
track
```

```python
# display the track on maps
from ipyleaflet import Map, Marker
m = Map(center=track[0].split(',')[::-1], zoom=15)
for t in track:
    mark = Marker(location=t.split(',')[::-1]) 
    m+=mark
m
```

```python
# list all start point
start_points = first_user[['start_location_y', 'start_location_x']]
m = Map(center=['31.280', '121.490'], zoom=10)
for index, row in start_points.iterrows():
    
    mark = Marker(location=[row['start_location_y'], row['start_location_x']])
    m+=mark
m
```

```python
# list all end_points
start_points = first_user[['end_location_y', 'end_location_x']]
m = Map(center=['31.280', '121.490'], zoom=10)
for index, row in start_points.iterrows():
    
    mark = Marker(location=[row['end_location_y'], row['end_location_x']])
    m+=mark
m
```

```python
%matplotlib inline
# 时间分布
first_user = first_user.sort_values(by='start_time')
first_user['hour'] = pd.to_datetime(first_user['start_time']).apply(lambda x: x.hour)
grouped = first_user.groupby('hour').count()

import matplotlib.pyplot as plt

plt.plot(grouped.index, grouped['orderid'])
plt.ylabel('number')
plt.xlabel('day time')
plt.show()
```
